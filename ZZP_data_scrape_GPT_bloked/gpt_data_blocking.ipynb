```python
import pandas as pd
import numpy as np

import pickle
import json
```

```python
dtypes = {'cfips': str}

# Load the data: AE_df.csv
AE_df = pd.read_csv('AE_df.csv', dtype=dtypes)

# Load the data: PCA_df.csv
PCA_df = pd.read_csv('PCA_df.csv', dtype=dtypes)

# Load the data: one_hot_df.csv
one_hot_df = pd.read_csv('one_hot_df.csv', dtype=dtypes)
```

```python
AE_np = AE_df.iloc[:, 1:].to_numpy()
PCA_np = PCA_df.iloc[:, 1:].to_numpy()
one_hot_np = one_hot_df.iloc[:, 1:].to_numpy()

print(AE_np.shape)
print(PCA_np.shape)
print(one_hot_np.shape)
```

```python
# concatenate the three arrays on the column axis
X = np.concatenate((AE_np, PCA_np, one_hot_np), axis=1)
X.shape
```

```python
data_np = np.repeat(X[np.newaxis, :, :], 47, axis=0)
data_np.shape
```

```python
# swap the axis so that the shape is (3142, 5, 47)
data_np = np.swapaxes(data_np, 1, 2)
data_np.shape
```
