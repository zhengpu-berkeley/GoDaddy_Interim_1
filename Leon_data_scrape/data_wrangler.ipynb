```python
# Create a DatetimeIndex for the 47 months starting from August 2018
date_index = pd.date_range(start='2019-08', periods=47, freq='M')

# Create a dataframe with the DatetimeIndex as the index
month_year = pd.DataFrame(index=date_index)

# Reset the index and rename the column to 'Date'
month_year = month_year.reset_index().rename(columns={'index': 'Date'})


# Convert the 'date' column to a pandas datetime object
month_year['Date'] = pd.to_datetime(month_year['Date'])



# Print the resulting DataFrame
display(month_year)

# Create a MultiIndex from the two DataFrames
index = pd.MultiIndex.from_product([month_year['Date'], start_data['cfips']], names=['Date', 'cfips'])

# Create a new DataFrame with all possible combinations of month-year and FIP code
base_data = pd.DataFrame(index=index).reset_index()
# Create new 'year' and 'month' columns by using the 'dt' accessor and the 'year' and 'month' attributes
base_data['year'] = base_data['Date'].dt.year
base_data['month'] = base_data['Date'].dt.month

# Left join base_data with start_data on fips column and add county_name and state columns from start_data
base_data = pd.merge(base_data.reset_index(), start_data[['cfips', 'county_name', 'state']], on='cfips', how='left')
display(base_data)
```

```python
#subtract fips2010 and cousub to get fips
#rename rent columns with year date

def rev_wrangle(df, year):

    cousub_lengths = df['cousub'].astype(str).apply(lambda x: len(x))

    df['fips'] = df['fips2010'].astype(str)
    fips = []
    for index, row in df.iterrows():
        
        
        fips.append(row['fips'][:-cousub_lengths[index]])
        
    df['fips'] = fips
    df['year'] = year

    
```

```python
rev_wrangle(rev_2018, 2018)
rev_wrangle(rev_2019, 2019)
rev_wrangle(rev_2020, 2020)
rev_wrangle(rev_2021, 2021)
rev_wrangle(rev_2023, 2023)
```
