```python
import pandas as pd
import numpy as np

df_join_4 = pd.read_csv('2d_data.csv')

# group the dataframe by the 'group' column
groups = df_join_4.groupby('fips')

# create a 3D numpy array from the groups
array_block = np.array([group_df.values for _, group_df in groups])
array_block = array_block.transpose((1,2,0))
# print the resulting 3D array
display(array_block.shape)
```

```python
two_d_df = pd.read_csv('2d_data.csv')
two_d_df.columns
```

```python
leon_cfips_list = array_block[0,2,:].tolist()
```

```python
# fill in the missing leading 0 for the fips
leon_cfips_list = [str(int(fips)).zfill(5) for fips in leon_cfips_list]
leon_cfips_list
```

```python
dtypes = {'cfips': str}
cfips_df = pd.read_csv('cfips.csv', dtype=dtypes)
```

```python
zzp_cfips_list = cfips_df['cfips'].tolist()
```

```python
data_block = np.ones((47, 34, 3142)) * np.nan
data_block = data_block.astype('object')

for i, fips in enumerate(leon_cfips_list):
    if fips == zzp_cfips_list[i-2]:
        data_block[:,:,i-2] = array_block[:,:,i-2]
    elif fips == zzp_cfips_list[i-1]:
        data_block[:,:,i-1] = array_block[:,:,i-1]
    elif fips == zzp_cfips_list[i]:
        data_block[:,:,i] = array_block[:,:,i]
    elif fips == zzp_cfips_list[i+1]:
        data_block[:,:,i+1] = array_block[:,:,i+1]
    elif fips == zzp_cfips_list[i+2]:
        data_block[:,:,i+2] = array_block[:,:,i+2]
    else:
        print(fips)
```
