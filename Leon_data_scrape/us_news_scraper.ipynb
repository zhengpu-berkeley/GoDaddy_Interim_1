```python


def split_columns(df, selected_indices, column_names, column):
    selected_cols = df[column].apply(lambda x: [x[i] for i in selected_indices if i < len(x)])
    selected_df = pd.DataFrame(selected_cols.tolist(), columns=column_names)
    return selected_df


```
