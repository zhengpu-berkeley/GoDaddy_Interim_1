### not much missing data, I would just impute with the geometric trend

```python
# impute the pct_bb_2020 with the pct_bb_2019 * (pct_bb_2019/pct_bb_2017)**(1/2), keep 1 decimal place
census_df['pct_bb_2020'] = census_df.apply(lambda x: round(x['pct_bb_2019'] * (x['pct_bb_2019']/x['pct_bb_2017'])**(1/2), 1) if np.isnan(x['pct_bb_2020']) else x['pct_bb_2020'], axis=1)
```

```python
# impute the pct_bb_2021 with the pct_bb_2020 * (pct_bb_2019/pct_bb_2017)**(1/2), keep 1 decimal place
census_df['pct_bb_2021'] = census_df.apply(lambda x: round(x['pct_bb_2020'] * (x['pct_bb_2019']/x['pct_bb_2017'])**(1/2), 1) if np.isnan(x['pct_bb_2021']) else x['pct_bb_2021'], axis=1)
```

```python
# impute the pct_college_2020 with the pct_college_2019 * (pct_college_2019/pct_college_2017)**(1/2), keep 1 decimal place
census_df['pct_college_2020'] = census_df.apply(lambda x: round(x['pct_college_2019'] * (x['pct_college_2019']/x['pct_college_2017'])**(1/2), 1) if np.isnan(x['pct_college_2020']) else x['pct_college_2020'], axis=1)

# impute the pct_college_2021 with the pct_college_2020 * (pct_college_2019/pct_college_2017)**(1/2), keep 1 decimal place
census_df['pct_college_2021'] = census_df.apply(lambda x: round(x['pct_college_2020'] * (x['pct_college_2019']/x['pct_college_2017'])**(1/2), 1) if np.isnan(x['pct_college_2021']) else x['pct_college_2021'], axis=1)

# impute the pct_foreign_born_2020 with the pct_foreign_born_2019 * (pct_foreign_born_2019/pct_foreign_born_2017)**(1/2), keep 1 decimal place
census_df['pct_foreign_born_2020'] = census_df.apply(lambda x: round(x['pct_foreign_born_2019'] * (x['pct_foreign_born_2019']/x['pct_foreign_born_2017'])**(1/2), 1) if np.isnan(x['pct_foreign_born_2020']) else x['pct_foreign_born_2020'], axis=1)

# impute the pct_foreign_born_2021 with the pct_foreign_born_2019 * (pct_foreign_born_2019/pct_foreign_born_2017), keep 1 decimal place
census_df['pct_foreign_born_2021'] = census_df.apply(lambda x: round(x['pct_foreign_born_2019'] * (x['pct_foreign_born_2019']/x['pct_foreign_born_2017']), 1) if np.isnan(x['pct_foreign_born_2021']) else x['pct_foreign_born_2021'], axis=1)

# impute the pct_it_workers_2018 with the pct_it_workers_2017 * (pct_it_workers_2019/pct_it_workers_2017)**(1/2), keep 1 decimal place
census_df['pct_it_workers_2018'] = census_df.apply(lambda x: round(x['pct_it_workers_2017'] * (x['pct_it_workers_2019']/x['pct_it_workers_2017'])**(1/2), 1) if np.isnan(x['pct_it_workers_2018']) else x['pct_it_workers_2018'], axis=1)

# impute the pct_it_workers_2020 with the pct_it_workers_2019 * (pct_it_workers_2019/pct_it_workers_2017)**(1/2), keep 1 decimal place
census_df['pct_it_workers_2020'] = census_df.apply(lambda x: round(x['pct_it_workers_2019'] * (x['pct_it_workers_2019']/x['pct_it_workers_2017'])**(1/2), 1) if np.isnan(x['pct_it_workers_2020']) else x['pct_it_workers_2020'], axis=1)

# impute the pct_it_workers_2021 with the pct_it_workers_2019 * (pct_it_workers_2019/pct_it_workers_2017), keep 1 decimal place
census_df['pct_it_workers_2021'] = census_df.apply(lambda x: round(x['pct_it_workers_2019'] * (x['pct_it_workers_2019']/x['pct_it_workers_2017']), 1) if np.isnan(x['pct_it_workers_2021']) else x['pct_it_workers_2021'], axis=1)

# impute the median_hh_inc_2018 with the median_hh_inc_2017 * (median_hh_inc_2019/median_hh_inc_2017)**(1/2), keep 1 decimal place
census_df['median_hh_inc_2018'] = census_df.apply(lambda x: round(x['median_hh_inc_2017'] * (x['median_hh_inc_2019']/x['median_hh_inc_2017'])**(1/2), 1) if np.isnan(x['median_hh_inc_2018']) else x['median_hh_inc_2018'], axis=1)

# impute the median_hh_inc_2020 with the median_hh_inc_2019 * (median_hh_inc_2019/median_hh_inc_2017)**(1/2), keep 1 decimal place
census_df['median_hh_inc_2020'] = census_df.apply(lambda x: round(x['median_hh_inc_2019'] * (x['median_hh_inc_2019']/x['median_hh_inc_2017'])**(1/2), 1) if np.isnan(x['median_hh_inc_2020']) else x['median_hh_inc_2020'], axis=1)

# impute the median_hh_inc_2021 with the median_hh_inc_2019 * (median_hh_inc_2019/median_hh_inc_2017), keep 1 decimal place
census_df['median_hh_inc_2021'] = census_df.apply(lambda x: round(x['median_hh_inc_2019'] * (x['median_hh_inc_2019']/x['median_hh_inc_2017']), 1) if np.isnan(x['median_hh_inc_2021']) else x['median_hh_inc_2021'], axis=1)
```
