```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dtypes = {'cfips': str}
data_df = pd.read_csv('county_area_population_density.csv', dtype=dtypes)
data_df.head()
```

```python
data_df['longitude'] = data_df['Geo Point'].apply(lambda x: float(x.split(',')[0]))
data_df['latitude'] = data_df['Geo Point'].apply(lambda x: float(x.split(',')[1]))
data_df.head()
```

```python
data_df['cfips'] = data_df['cfips'].astype(str)
data_df.head()
```

```python
data_df['cfips'] = data_df['cfips'].apply(lambda x: x[:-2])
data_df.head()
```

```python
cfips_df = pd.read_csv('cfips.csv', dtype=dtypes)
```

```python
# merge the longitude and latitude into the cfips dataframe
new_cfips_df = cfips_df.merge(data_df[['cfips', 'longitude', 'latitude']], on='cfips', how='left')
new_cfips_df
```

```python
# show the counties that don't have longitude and latitude
new_cfips_df[new_cfips_df['longitude'].isna()]
```

```python
new_cfips_df = new_cfips_df.merge(data_df[['cfips', 'ALAND', 'Population per square meter']], on='cfips', how='left')
new_cfips_df
```

```python
# change the name of the column
new_cfips_df = new_cfips_df.rename(columns={'ALAND': 'area'})
new_cfips_df = new_cfips_df.rename(columns={'Population per square meter': 'pop_per_sq_meter'})
new_cfips_df
```

```python
# save the new cfips dataframe
new_cfips_df.to_csv('longi_latti_area_pop_persq_meter.csv', index=False)
```
